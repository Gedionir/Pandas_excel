```python
import pandas as pd
```

```python
#leitura de arquivos
df1 = pd.read_excel("/content/drive/MyDrive/Datasets/Aracaju.xlsx")
df2 = pd.read_excel("/content/drive/MyDrive/Datasets/Fortaleza.xlsx")
df3 = pd.read_excel("/content/drive/MyDrive/Datasets/Natal.xlsx")
df4 = pd.read_excel("/content/drive/MyDrive/Datasets/Recife.xlsx")
df5 = pd.read_excel("/content/drive/MyDrive/Datasets/Salvador.xlsx")
```

```python
#Juntando todos os arquivos
df = pd.concat([df1,df2,df3,df4,df5])
df.head()
```

```python
#alterando o tipo de dado da coluna LojaID, pois não vai precisar fazer calculos com os dados dessa coluna.
df["LojaID"] = df["LojaID"].astype("object")
```

```python
#Consultando linhas com valores faltantes
df.isnull().sum()
```

```python
#Subistituir valor nulos pela Média
#inplace - Seja modificado em memória
#Subistituindo nulos por 0 = df["Vendas"].fillna(0, inplace=True)
#Removendo linha nulas = df.dropna(inplace=True)
#Removendo linhas com base somente a colunas vendas = df.dropna(subset=["Vendas"], inplace=True)
#Removendo somente linhas que tem valos nulos das colunas = df.dropna(how="all", inplace=True)
df["Vendas"].fillna(df["Vendas"].mean(), inplace=True)
```

```python
df.isnull().sum()
```

```python
df.sample(15)
```

```python
df["Vendas"].mean()
```

```python
#substituindo os valores nulos por zero
df["Vendas"].fillna(0, inplace=True)
```

```python
#Apagando as linhas com valores nulos
df.dropna(how="all", inplace=True)
```

```python
#apagando as linhas com valores nulos com base em primeira coluna
df.dropna(subset=["Vendas"], inplace=True)
```

```python
#Remover linhas que estejam com valores falatntes em todas as colunas
df.dropna(how="all", inplace=True)
```

**CRIANDO COLUNAS NOVAS**

```python
#Criando nova coluna multiplicando quantidade por valor da venda
df["Receita"] = df["Vendas"].mul(df["Qtde"])
df.head()
```

```python
df["Receita/Vendas"] = df["Receita"] / df['Vendas']
```

```python
df.head()
```

```python
#retornando a maior receita
df["Receita"].max()
```

```python
#retornando a menor receita
df["Receita"].min()
```

```python
#Retorna o top 3 com base na coluna receita
#nlargest
df.nlargest(3, "Receita")
```

```python
#Retorna o top 3 piores receitas com base na coluna receita
#nlargest
df.nsmallest(3, "Receita")
```

```python
#Agrupamento por cidade
df.groupby("Cidade")["Receita"].sum()
```

```python
#Ordenado um conjunto de dados -   'ascending=False ' do maior para o menor
df.sort_values("Receita", ascending=False).head(10)
```

**TRABALHANDO COM DATAS**

```python
df.dtypes
```

```python
#Transformando colunas de datetime e inteiros
df["Data"] = df["Data"].astype("int64")
```

```python
#Transformado coluna de data em tipo datetime
df["Data"] = pd.to_datetime(df["Data"])
```

```python
df.dtypes
```

```python
#Agrupando por data por ano
#dt - datetime
#dt.year - vai na minha data e extrai apenas o ano
df.groupby(df["Data"].dt.year)["Receita"].sum()
```

```python
#Criando coluna Ano_venda
df["Ano_Venda"] = df["Data"].dt.year
```

```python
df.sample(5)
```

```python
#Extraíndo mês e o dia
df["Mes_Venda"], df["Dia_Venda"] = (df["Data"].dt.month, df["Data"].dt.day)
```

```python
df.sample(5)
```

```python
#Data mais antiga
df["Data"].min()
```

```python
#Calculando diferença de dias entre datas proximas menos anterior
df["Diferenca_Dias"] = df["Data"] - df["Data"].min()
```

```python
df.sample(5)
```

```python
#trimestre venda
df["Trimestre_Venda"] = df["Data"].dt.quarter
```

```python
df.sample
```

```python
#filtrando as vendas de 2019 do mês 4
venda_marco_19 = df.loc[(df["Data"].dt.year == 2019) & (df["Data"].dt.month == 3)]
```
